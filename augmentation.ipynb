# Data Transformations with scikit-image

For scikit-image documentation: https://scikit-image.org/

Images from Columbia University Image Library: https://www.tensorflow.org/datasets/catalog/coil100

```python
import os
import skimage as sk

from skimage import transform
from skimage import util
from skimage import io
from skimage import exposure
from skimage.color import rgb2gray
```

### Display the image

```python
glue = io.imread('coil-100/obj33__335.png')
plt.imshow(glue)
```

### Rotation

```python
# degree can be a value between -180 and 180
degree = 90
glue_rotate = sk.transform.rotate(glue, degree)
plt.imshow(glue_rotate)
```

### Add Noise

```python
# mode can be ‘gaussian’, ‘localvar’, ‘poisson’, ‘salt’, ‘pepper’, ‘s&p’, ‘speckle’
mode = 'gaussian'
glue_noise = sk.util.random_noise(glue, mode=mode)
plt.imshow(glue_noise)
```

### Horizontal Flip

```python
glue_horiz_flip = glue[:, ::-1]
plt.imshow(glue_horiz_flip)
```

### Grayscale

```python
glue_gray = rgb2gray(glue)
plt.imshow(glue_gray, cmap='gray')
```

### Color inversion

```python
glue_invert = util.invert(glue)
plt.imshow(glue_invert)
```
